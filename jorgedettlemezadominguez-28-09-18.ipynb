```python
import pylab as pl
import numpy as np

def heun(f,x,y,h,m):
    u=np.zeros([m,2], dtype=float)
    for i in range(m):
        yn=y+h*f(x,y)
        y=y+h/2*(f(x,y)+f(x+h,yn))
        x=x+h
        u[i,0]=x
        u[i,1]=y
    return u
def f(x,y):
    return y-x**2 +x+1
u=heun(f,0,1,0.1,20)

pl.plot(u[:,0],u[:,1],'or')
def y(x):
    return np.exp(x) + x**2 + x
x=pl.arange(0,2.1,0.1)
pl.plot(x,y(x),'o--')
pl.grid(True)
pl.show()
```

```python
import pylab as pl
import numpy as np

def rk2(f,x,y,h,m):
    u=np.zeros([m,2], float)
    for i in range(m):
        k1=h*f(x,y)
        k2=h*(f(x+h,y+k1))
        x=x+h
        y=y+0.5*(k1+k2)
        u[i,0]=x
        u[i,1]=y
    return u
def f(x,y):
    return y-x**2 +x+1
u=rk2(f,0,1,0.1,20)

pl.plot(u[:,0],u[:,1],'or')
def y(x):
    return np.exp(x) + x**2 +x
x=pl.arange(0,2.1,0.1)
pl.plot(x,y(x),'b--')
pl.grid(True)
pl.show()
```
